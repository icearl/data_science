```python
# 导入库
import matplotlib.pyplot as plt
import numpy as np
```

## 1. 画一个（x, y）曲线

```python
# 定义数据
x = np.linspace(-1, 1, 50)
y = 2*x + 1
```

```python
plt.figure()   # 定义一个图像窗口
plt.plot(x, y) # 用plt.plot画(x ,y)曲线
plt.show()     # 显示图像
```

## 2. 画两个曲线

```python
# 准备数据
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
```

```python
plt.figure(num=3, figsize=(8, 5),) # 编号为3；大小为(8, 5)
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.show()
```

## 3.  设置坐标轴

```python
plt.figure(num=3, figsize=(8, 5),) # 编号为3；大小为(8, 5)
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-1, 2))   # 设置 坐标轴范围
plt.ylim((-2, 3))
plt.xlabel('I am x')  # 设置坐标轴名称
plt.ylabel('I am y')
plt.show()
```

```python
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)  # 设置坐标轴刻度
plt.yticks([-2, -1.8, -1, 1.22, 3],[r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])
plt.show()
```

```python
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks) 
plt.yticks([-2, -1.8, -1, 1.22, 3],[r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])
# 用plt.gca获取当前坐标轴信息. 使用.spines设置边框：右侧边框；
# 使用.set_color设置边框颜色：默认白色； 
# 使用.spines设置边框：上边框；
# 使用.set_color设置边框颜色：默认白色；
ax = plt.gca()
ax.spines['right'].set_color('red')
ax.spines['top'].set_color('blue')
plt.show()
```

```python
plt.figure(num=4, figsize=(8, 5),) # 编号为3；大小为(8, 5)
# 设置坐标轴位置
ax = plt.gca()
# 使用.xaxis.set_ticks_position设置x坐标刻度数字或名称的位置：bottom.（所有位置：top，bottom，both，default，none）
# 同理，使用.yaxis.set_ticks_position设置y坐标刻度数字或名称的位置：left.（所有位置：left，right，both，default，none）
ax.xaxis.set_ticks_position('none')
# 使用.spines设置边框：x轴；使用.set_position设置边框位置：y=0的位置；（位置所有属性：outward，axes，data）
# 同理，使用.spines设置边框：y轴；使用.set_position设置边框位置：x=0的位置；（位置所有属性：outward，axes，data） 使用plt.show显示图像.
ax.spines['bottom'].set_position(('data', 4))
plt.plot(x, y2)
plt.show()
```

## 4. 添加图例

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
#set x limits
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# set new sticks
new_sticks = np.linspace(-1, 2, 5)
plt.xticks(new_sticks)
# set tick labels
plt.yticks([-2, -1.8, -1, 1.22, 3],
           [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$really\ good$'])

# set line syles
# 这里可以添加图例
l1, = plt.plot(x, y1, label='linear line')
l2, = plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--', label='square line')
# plt.legend(loc='upper right')
```

```python
# 调整图例
# 如果我们想单独修改之前的 label 信息, 给不同类型的线条设置图例信息. 
# 我们可以在 plt.legend 输入更多参数. 
# 如果以下面这种形式添加 legend, 我们需要确保, 
# 在上面的代码 plt.plot(x, y2, label='linear line') 和 plt.plot(x, y1, label='square line') 中有用变量 l1 和 l2 分别存储起来.
# 而且需要注意的是 l1, l2,要以逗号结尾, 因为plt.plot() 返回的是一个列表.
l1, = plt.plot(x, y1, label='linear line')
l2, = plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--', label='square line')
# plt.legend(handles=[l1, l2], labels=['up', 'down'],  loc='best')
plt.legend(handles=[l1, l2], loc='best')
```

```python
"""
其他 loc 参数
其中’loc’参数有多种，’best’表示自动分配最佳位置，其余的如下：

 'best' : 0,          
 'upper right'  : 1,
 'upper left'   : 2,
 'lower left'   : 3,
 'lower right'  : 4,
 'right'        : 5,
 'center left'  : 6,
 'center right' : 7,
 'lower center' : 8,
 'upper center' : 9,
 'center'       : 10,
 """
```
